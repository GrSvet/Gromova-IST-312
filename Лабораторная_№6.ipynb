```python
import matplotlib.pyplot as plt
fahrenheit = [[-67.0], [-34.0], [0], [34.0], [54.0], [67.0], [100]]
kelvin = [[218.15], [236.48], [255.37], [274.26], [285.37], [292.59], [310.92]]
```

```python
plt.figure(figsize=(15,8), dpi=60)
plt.scatter(fahrenheit, kelvin, label='входные данные', color='green', marker='$f$');
plt.xlabel('fahrenheit')
plt.ylabel('kelvin')
plt.legend()
plt.grid(True)
plt.show()
```

```python
for c, f in zip(fahrenheit, kelvin):
  print(f'фаренгейт{c}= кельвин {f}') 
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(fahrenheit, kelvin)
lr.predict([[256], [123]])
lr.coef_
lr.intercept_
fahrenheit_test = [[-50], [10], [30], [20], [10], [70], [87]]
kelvin_test = lr.predict(fahrenheit_test)
kelvin_test
```
